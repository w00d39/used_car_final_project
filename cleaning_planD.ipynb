```python
import pandas as pd
import sklearn
import time

```

```python
start_time = time.time()  # Record the start time

# Process the file in chunks
chunk_size = 10000  # Number of rows per chunk
all_chunks = []  # List to store chunks

for chunk in pd.read_csv('used_cars_data.csv', chunksize=chunk_size):
    all_chunks.append(chunk)  # Append each chunk to the list

# Combine all chunks into a single DataFrame
data = pd.concat(all_chunks, ignore_index=True)

end_time = time.time()  # Record the end time
elapsed_time = end_time - start_time  # Calculate elapsed time
print(f"Elapsed time: {elapsed_time:.2f} seconds")

# Now you can clean the data using the `data` DataFrame
print(data.head())  # Example: Print the first few rows of the combined DataFrame
```

```python
#drop columns
#combine_fuel_economy
#is_certified
#main_picture_url
#theft_title
#vehicle_damage_category
useless_columns = ['combine_fuel_economy', 'is_certified', 'main_picture_url', 'theft_title', 'vehicle_damage_category']
data.drop(columns=useless_columns, inplace=True)
```
